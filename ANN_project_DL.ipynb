```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten
from tensorflow.keras.datasets import fashion_mnist
from tensorflow.keras.utils import to_categorical
```

```python
#Load Dataset
(X_train, y_train), (X_test, y_test) = fashion_mnist.load_data()
```

```python
#Check Dataset Shape
print(X_train.shape)
print(X_test.shape)
```

```python
print(y_train.shape)
print(y_test.shape)
```

```python
#Normalize Data
X_train = X_train / 255.0
X_test = X_test / 255.0
```

```python
#Convert Target to Categorical
y_train = to_categorical(y_train)
y_test = to_categorical(y_test)

```

```python
print(y_train.shape)
print(y_test.shape)
```

```python
#Build ANN Model
#Flatten ---- Converts 28×28 → 784 ---- (784,)
#Dense(128) ---- Hidden layer ---- 128 neurons
#Dense(64) ---- Hidden layer ---- 64 neurons
#Dense(10) ---- Output layer ---- 10 classes
```

```python
model = Sequential()

model.add(Flatten(input_shape=(28, 28)))
model.add(Dense(128, activation='relu'))
model.add(Dense(64, activation='relu'))
model.add(Dense(10, activation='softmax'))
```

```python
# Compile Model
model.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)
```

```python
history = model.fit(
    X_train,
    y_train,
    epochs=10,
    batch_size=32,
    validation_data=(X_test, y_test)
)
```

```python
#Evaluate Model
loss, accuracy = model.evaluate(X_test, y_test)
print("Test Accuracy:", accuracy)
```

```python
#Make Predictions
predictions = model.predict(X_test)
```

```python
index = 1000
plt.imshow(X_test[index])
plt.show()

print("Predicted Class:", np.argmax(predictions[index]))
print("Actual Class:", np.argmax(y_test[index]))
```
